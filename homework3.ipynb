5-1 for~in 리터럴

```python
def addall(list):
    sum = 0
    for i in list:
        sum += i
    return sum

print(addall([1]))
print(addall([1,2,3,4,5,6,7,8,9]))  
```

정수의 합 sum을 0으로 초기화하고, for~in 리터널을 통해 결과값 출력

5-1 리스트 내포 방식

```python
def addall(list):
    return sum([i for i in list])

print(addall([1]))
print(addall([1,2,3,4,5,6,7,8,9]))
```

sum 이라는 내장함수를 통해 쉽게 구할 수 있음

5-2 for~in 리터럴

```python
def addallodd(list):
    sum = 0
    l = []
    for i in list:
        if(i%2==1): l.append(i)
    for odd in list:
        sum += odd
    return sum

print(addall([1]))
print(addall([1,2,3,4,5,6,7,8,9]))
```

리스트에 홀수를 추가 후 sum한 값을 return한다.

5-2 리스트 내포 방식

```python
def addallodd(list):
    return sum([i for i in list if i%2==1])

print(addall([1]))
print(addall([1,2,3,4,5,6,7,8,9]))
```

리스트 내포 방식을 이용하면 위 for~in 리터럴방식과 비교했을 때  
훨씬 간단하게 표현할 수 있다.

7-1 키 오름차순 정렬

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
l = list(zip(d.keys(), d.values()))
m = sorted(l, reverse=False)
for k in m:
    print(k)
```

키(key)기준으로 정렬 reverse=False 는 오름차순

7-2 키 내림차순 정렬

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
l = list(zip(d.keys(), d.values()))
m = sorted(l, reverse=True)
for k in m:
    print(k)
```

내림차순으로 정렬할려면 reverse=True으로 바꾸면 됨.

7-3 값 오름차순 정렬

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
l = list(zip(d.keys(), d.values()))
l.sort(key = lambda x:x[1])
for k in l:
    print(k)
```

x[1] => 인덱스가 1인 원소로 오름차순 정렬

7-4 값 내림차순 정렬

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
l = list(zip(d.keys(), d.values()))
l.sort(key = lambda x:-x[1])
for k in l:
    print(k)
```

-x[1] => 인덱스가 1인 원소로 내림차순 정렬

```python
def isStrange(num):
    m = len(str(num))
    if(m>1 and num%m==0): return isStrange(num//m)
    elif(num>=0 and m==1): return True
    else: return False

print(isStrange(0))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))

```

맨처음 if문을 통해 음수인 경우를 걸러내고, 그 다음 num이 한자리면 Strange이므로 무조건 True를 반환한다.  
num이 두 자리 이상인 경우 자릿수로 나눠서떨어지면 재귀함수 호출한다.  
계속 자릿수로 나눠서 한자리가 되면 True이고 중간에 나누어 떨어지지 않는경우 False가 나온다.

```python
def solve(a):
    sum = 0
    l = []
    for i in a:
        sum += i
        l.append(sum)

    return([i for i in l if sum % i ==0 and i>=max(a)])

print(solve([1,2,1,1,1,2,1,3]))
print(solve([1,1,1,1]))
print(solve([9999] * 10**5))
```

a 리스트 내 누적합의 경우들을 모두 l 리스트에 넣은다음  
버스 사이즈는 모든 팀원 수의 합의 약수 중 하나를 가질 수 있으므로 sum % i ==0  
그리고 같은 그룹에 속한 팀원들은 분리되면 안되므로 버스 사이즈는 항상  
각 그룹 팀원 수보다는 항상 같거나 커야하므로 i>=max(a) 조건이 필요하다

```python
import requests
import re

req1 = requests.get('https://naver.com')
source = req1.text

body = re.search('<body.*</body>', source, re.I|re.S)

body = body.group()
body = re.sub('<script.*?>.*?</script>', '', body, 0, re.I|re.S)

req11 = re.sub('<.+?>', '', body, 0, re.I|re.S)

req12 = re.sub('&nbsp; |\t|\r|\n', '', req11)
pattern = re.compile(u'<\/?\w+\s*[^>]*?\/?>', re.DOTALL | re.MULTILINE | re.IGNORECASE | re.UNICODE)
req13 = pattern.sub(u"", req12)

req14 = (req13.split(' '))
req15 = ' '.join(req14).split()


i = ' '.join(req15)
text = re.sub('[-=+,#/\?:;}_{^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', i)
text1 = re.sub('[a-zA-z]','',text)

word_dic = {}

for word in text1.split():
    if word in word_dic:
        word_dic[word] += 1
    else:
        word_dic[word] = 1

print(word_dic)
print ('req1.text = ', len(req14))
```

word_dic = {} 사전을 하나 만들고, for문을 통해 사전에 등록되어있으면 1씩올려주고 없으면 1로 등록
